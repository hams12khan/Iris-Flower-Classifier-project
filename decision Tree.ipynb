```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 

#this is ou model from decison TreeClassifier
from sklearn.tree import DecisionTreeClassifier

#here we are loading the dataset iris
from sklearn.datasets import load_iris

#this is for drawing the decision tree that we creating
from sklearn import tree
```

```python
df=load_iris()
df
```

```python
dataset=pd.DataFrame(df.data,columns=df.feature_names)
dataset["target"]=df.target
dataset
```

```python
#this is our model Decision tree Classifier
model=DecisionTreeClassifier()
model.fit(dataset.iloc[:,0:4],dataset.iloc[:,4:5])

#here we are taking the picture of tree that was create
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

```python
print("Enter the Parameter of flower")
print("Enter sepal parameter")
sepal_length,sepal_width=map(float,input().split())
print("Enter the petal Width")
petal_length,petal_width=map(float,input().split())
y_pred=model.predict([[sepal_length,sepal_width,petal_length,petal_width]])
if y_pred==0:
    print("setosa")
elif y_pred==1:
    print("versicolor")
else:
    print("virginica")
```
